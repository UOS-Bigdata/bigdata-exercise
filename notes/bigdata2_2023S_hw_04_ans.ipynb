## 1. 확률과 함수 (총 10점)

다음과 같은 게임을 생각해 보자.

- 4개의 주사위를 던져서 얻은 4개의 결과에서 1의 눈이 최소한 한 번 나오면 승리한다.


1의 눈이 최소한 한 번 나오는 사건을 $A$ 라고 하면 A의 여사건 $A^c$은  어떤 사건인가요?

$A^c$ 는 4개의 주사위를 던지는 경우 1의 눈이 한 번도 나오지 않는 사건입니다. 이제 $A^c$ 의 확률은 어떻게 계산할까요? 

하나의 주사위를 던지는 경우 1의 눈이 아닐 사건의 확률은 5/6입니다. 4개의 주사위를 던지는 것은 독립시행이므로 확률은 곱셈 법칙을 이용합니다.



$$ P(A^c) = \left ( \frac{5}{6} \right ) \left ( \frac{5}{6} \right ) \left ( \frac{5}{6} \right ) \left ( \frac{5}{6} \right ) = 0.482253$$


````python
(5/6)**4
````

여사건의 법칙 $P(A) = 1 - P(A^c)$ 을 적용하면 1의 눈이 최소한 한 번 나오는 사건의 확률은 다음과 같습니다. 

````python
1-(5/6)**4
````

$$ S = \{ 1, 2, 3, 4, 5 ,6 \} $$

````python
def count_one(x):
  total = 0
  for value in x:
    if value == 1:
      total = total + 1
  return(total) 
````

**문제 3** [3점]

3개의 주사위를 던졌을 때 나온 숫자들의 합을 생각하자. 문제 1번에서 정의한 표본공간 `dice_3` 를 인자로 받아서
3개의 주사위를 던졌을 때 숫자들의 합에 대한 확률 분포을 구하는 함수 `dist_sum_roll`을 작성하시오. 

함수를 작성할 때 `dist_sum_roll(dice_3)` 의 결과는 2개의 열을 가지는 데이터프레임이다. 
- 첫 번째 열은 가능한 합의 값을 가진다. 
- 두 번째 열은 합의 값에 대한 확률을 나타낸다. 

예를 들면 다음과 같은 형식으로 결과가 나와야 한다.

| value | prob |
|-----|------|
|  3  | 0.004630 |
|  4  | 0.013889 |
| ... | ...|
| 18 | 0.004630|

아래 힌트를 참조하여 코드를 작성하세요.

````python
def dist_sum_roll(x):
    outcomes = []
    n = len(x)
    for roll in x:
        tsum = np.sum(roll)
        outcomes.append(tsum)
    p = [ 1/n for i in np.arange(n) ]
    
    df_dist_all = pd.DataFrame({'outcomes': outcomes, 'p':p})
    df_dist = df_dist_all.groupby(by='outcomes').sum().reset_index()
   
    return df_dist
````

[힌트]
- 3개의 주사위를 던졌을 때 나타나는 결과의 확률은 모두 동일하게 1/216 으로 동일한다. 예를 들어
$$ P([1,1,1]) = P([1,1,12)= \cdots = P([6,6,6]) = \frac{1}{6} \times \frac{1}{6} \times \frac{1}{6}  = \frac{1}{216} $$
- 리스트의 길이는 함수 `len` 을 사용하여 구할 수 있다.
- 한 개의 문자 또는 숫자를 n 번 반복값들을 가지는 리스트를 만드는 방법은 다음과 같다.   
```
[x for i in np.arange(n)]
```
- 다른 방법으로 `numpy` 라이브러리의 `repeat` 함수를 사용하여 문자나 숫자가 반복된 벡터를 만들 수 있다.
```
np.repeat( x, n)
```

**(4) 쉼표를 포함하는 숫자로 구성된 문자열에서 쉼표 없애기**


이제 쉼표를 포함하는 숫자로 구성된 문자열에서 쉼표 없애보자.

```
'6,076' ->> '6076'
```

이러한 작업을 하기 위하여 앞에서 데이터프레임에 적용되는 메소드 `replace(old,new)` 를 적용하는데 문제가 있다. 

- **데이터프레임에 적용되는** 메소드 `replace(old,new)` 는 셀의 내용이 정확하게 `old` 와 일치하는 경우만 `new` 로 바꾸어 준다. 따라서 쉼표를 포함한 숫자의 형태가 너무 다양하기 때문에 적용할 수가 없다.

- 하지만 **문자열에 적용되는** 메소드 `replace(old,new)` 는 문자열의 각 문자를 모두 찾아 바꾸어 줄 수 있다. 

다음에 예제에서 같은 메소드 `replace(old,new)` 가 적용되는 대상에 달라지면 기능과 결과가 다른 것을 알  수 있다.


````python
ex1 = pd.DataFrame({"0-9세": ["6,076",	"8,764"], "10-19세":["10918","11,504"]})
ex1
````

````python
ex1.replace(",", "")
````

````python
ex1.replace("6,076", "6076")
````

````python
ex1.iloc[0,0]
````

````python
ex1.iloc[0,0].replace(",","")
````

이제 문자열을 인자로 받아서 쉼표 `,` 를 공백이 없는 문자(`""`)로 바꾸어 주는 함수 `replace_comma2null` 를 다음과 같이 정의하자.

- 아래에서 `isinstance(s, str)`는 `s` 가 문자열이면 `True`가 된다.

````python
def replace_comma2null(s):
  if isinstance(s, str):
    res = s.replace("," ,"")
  else:
    res = s
    
  return res
````

````python
replace_comma2null(ex1.iloc[0,0])
````

이제 우리가 배운 `applymap` 메소드를 이용하여 함수 `replace_comma2null`를  데이터프레임의 모든 셀에 적용해 보자. 

````python
df4 = df3.applymap(replace_comma2null)
df4
````

- `pd.to_datetime( series, format)` 함수는 문자가 `format`으로 표시되어 있는 시리즈(Series)를  날짜 형식을 가지는 시리즈 자료로 만들어 준다.

  + 문자열이 `2020-12-01` 로 표시된 `format` 은 `%Y-%m-%d` 로 나타낸다.
  + 다양한 날짜 형식에 대한 문법은 [여기](https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior)를 참조하자. 

-  `dt.month` 는 날짜 형식을 가지는 시리즈 자료에서 월(month) 를 출력하는 메소드이다.


````python
df5['일자'] = pd.to_datetime(df4['일자'], format='%Y-%m-%d')
df5['월'] = df5['일자'].dt.month
````

**문제 3**

먼저, 다음과 같이 날짜 형식의 자료에서 연도를 구하는 메소드는 `dt.year` 이다.

````python
df5['일자'].dt.year
````

다음으로, 데이터프레임 `final`의 복사본 `final_copy` 를 만들고 각 행에 `연도` 와 총 확진자수 `total` 열을 추가하자.

````python
# 데이터프레임의 복사본 만들기
final_copy = final.copy()  
# 연도를 나타내는 열을 추가하기
final_copy.loc[:,'연도'] = final_copy['일자'].dt.year
# 확진자의 합계를 계산하여 새로운 열로 추가하기 
final_copy.loc[:,'total'] = final_copy.drop(columns = ["일자", "월","연도"]).sum(axis=1)
````

위에서 만든 데이터프레임 `final_copy` 을 이용하여 **연도별, 월별 총 확진자수**를 구하는 코드를 작성하시오. 

````python
final_copy.loc[:,['연도','월','total']].groupby(by=['연도','월'] ).sum()
````
